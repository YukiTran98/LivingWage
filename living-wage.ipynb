# Data Visualization

**What we can do with this dataset?**

- 5 highest ranked cities in terms of total population
- 5 cities with the highest population density
- 5 cities with the highest population growth
- Top cities with the highest/lowest avg_living_wages

```python
import pandasql as psql
```

```python
# Top 5 City with highest pupolation in 2020
rank_city = psql.sqldf("select rank_2020, city from df order by rank_2020 ASC limit 5")
rank_city
```

```python
# Top 5 city with the highest population growth 
population_increase_city = psql.sqldf("""Select city, (population_2020-population_2010)/10 as increase_per_year 
                                from df
                                order by (population_2020-population_2010)/10 DESC
                                limit 5""")
population_increase_city
```

```python
plt.figure(figsize=(8, 4), dpi=120)
plt.xticks(fontsize=9, rotation=45)
plt.yticks(fontsize=9)
plt.title('Top 5 Cities Population Increased from 2010 to 2020')
plt.bar(population_increase_city.city,population_increase_city.increase_per_year)
plt.xlabel('Cities')
plt.ylabel('Population increase per year')
plt.grid(True,linestyle="--", color='gray')
plt.show()
```

```python
# Top 5 cities with the highest population density
top_density = psql.sqldf("""select city, density
                            from df
                            order by density DESC
                            limit 5""")
top_density
```

```python
plt.figure(figsize=(8, 4), dpi=120)
plt.xticks(fontsize=9, rotation=45)
plt.yticks(fontsize=9)
plt.title('Top 5 Cities Population Increased from 2010 to 2020')
sns.barplot(x=top_density.city, y = top_density.density)
plt.xlabel('Cities')
plt.ylabel('Density')
plt.grid(True,linestyle="--", color='gray')
plt.show()
```

**Note** : Because there are many subgroups of the population with different living wages, in this analysis I will just look at the average living wages across all groups.

```python
# Adding a new column to calculate the average living wage for all groups
df['avg_living_wage'] = df.loc[:,'one_adult_no_kids_living_wage':'two_adults_both_working_three_kids_living_wage'].mean(axis=1)
df['avg_living_wage']=round(df['avg_living_wage'],2)
df.head()
```

```python
# Top 5 city with highest avg_living_wage
top_avg_living_wage = psql.sqldf("""select city, avg_living_wage
                                    from df
                                    order by avg_living_wage DESC
                                    limit 5""")
top_avg_living_wage
```

```python
plt.figure(figsize=(8, 4), dpi=120)
plt.xticks(fontsize=9, rotation=45)
plt.yticks(fontsize=9)
plt.title('Top 5 city with highest avg_living_wage')
sns.barplot(x=top_avg_living_wage.city, y = top_avg_living_wage.avg_living_wage)
plt.xlabel('Cities')
plt.ylabel('avg_living_wage')
plt.grid(True,linestyle="--", color='gray')
plt.show()
```

```python
# Top 5 city with lowest avg_living_wage
bottom_avg_living_wage = psql.sqldf("""select city, avg_living_wage
                                    from df
                                    order by avg_living_wage ASC
                                    limit 5""")
bottom_avg_living_wage
```

```python
plt.figure(figsize=(8, 4), dpi=120)
plt.xticks(fontsize=9, rotation=45)
plt.yticks(fontsize=9)
plt.title('Top 5 city with lowest avg_living_wage')
sns.barplot(x=bottom_avg_living_wage.city, y = bottom_avg_living_wage.avg_living_wage)
plt.xlabel('Cities')
plt.ylabel('avg_living_wage')
plt.grid(True,linestyle="--", color='gray')
plt.show()
```

```python
# Examining the link between population and living wages
plt.figure(figsize=(8, 4), dpi=120)
plt.xlabel('Population')
plt.ylabel('avg_living_wage')
plt.title('Relationship between Population and avg_living_wage')
plt.scatter(x=df.population_2020, y = df.avg_living_wage)
plt.show()
```

**Note** : It can be seen that the average living wage in each city does not seem to have much to do with the total population in that city.

```python
# Examining the link between density and living wages
plt.figure(figsize=(8, 4), dpi=120)
plt.xlabel('Density')
plt.ylabel('avg_living_wage')
plt.title('Relationship between Population and avg_living_wage')
plt.scatter(x=df.density, y = df.avg_living_wage)
plt.show()
```

**Note** : There seems to be a slight positive correlation between population density and the average living standard of cities.

# Clustering

```python
df1=df.drop(['city','state'],axis=1)
```

```python
df1.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
sc = StandardScaler()
cl_sc=sc.fit_transform(df1)
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit(cl_sc)
```

```python
df1["clustering"] = clusters.labels_
df1.head()
```

```python
df1['city']=df.city
```

```python
df1.groupby('clustering').mean()
```

```python
pre = cl_sc
pca = PCA(n_components=2) 
pca.fit(pre)
pca_pre = pca.transform(pre)
pca_data = pd.DataFrame(pca_pre)
```

```python
for i in df1["clustering"].unique():
    tmp = pca_data.loc[df1["clustering"] == i]
    plt.scatter(tmp[0], tmp[1], label=i)
plt.legend()
```

**Note** : 

* Cluster 0: High populatuon and density, and most expensive to live, least population grow
* Cluster 1: Low density and least expensive to live
* Cluster 2: Middle density and less enpensive to live
* Cluster 3: Middle density and more expensive to live, most population grow
